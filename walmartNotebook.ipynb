```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ejemplo de DF para total de ventas por mes
data_ventas_mes = {
    'mes': ['2023-01', '2023-02', '2023-03', '2023-04', '2023-05'],
    'total_ventas': [15000, 20000, 25000, 30000, 35000]
}
df_ventas_mes = pd.DataFrame(data_ventas_mes)

# Ejemplo de DF para total de productos por categoría
data_productos_categoria = {
    'categoria': ['Electrónica', 'Ropa', 'Alimentos', 'Hogar'],
    'total_productos': [50, 100, 200, 30]
}
df_productos_categoria = pd.DataFrame(data_productos_categoria)

# Ejemplo de DF para total de ventas por empleado
data_ventas_empleado = {
    'empleado': ['Juan Pérez', 'Ana Gómez', 'Luis Martínez'],
    'total_ventas': [50000, 70000, 30000]
}
df_ventas_empleado = pd.DataFrame(data_ventas_empleado)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_ventas_mes, x='mes', y='total_ventas', palette='viridis')
plt.title('Total de Ventas por Mes')
plt.xlabel('Mes')
plt.ylabel('Total de Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_productos_categoria, x='categoria', y='total_productos', palette='pastel')
plt.title('Total de Productos por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Total de Productos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_ventas_empleado, x='empleado', y='total_ventas', palette='coolwarm')
plt.title('Total de Ventas por Empleado')
plt.xlabel('Empleado')
plt.ylabel('Total de Ventas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
